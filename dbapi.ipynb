## Edit

```python
edit_1 = pd.read_csv("E:/study/aml/appaisal project/edit_1.csv", encoding = "ISO-8859-1")
edit_2 = pd.read_csv("E:/study/aml/appaisal project/edit_2.csv", encoding = "ISO-8859-1")
edit_3 = pd.read_csv("E:/study/aml/appaisal project/edit_3.csv", encoding = "ISO-8859-1")
```

```python
edit_merge = pd.concat([edit_1,edit_2,edit_3])
cols = ["pageid","eid","uid","timestamp","flags","sizechange","comment","diff","inline"]
edit_merge = edit_merge[cols]
```

```python
edit_merge.to_csv("E:/study/aml/appaisal project/edit_merge.csv",index=False, header=True)
```

```python
edit_merge
```

```python
edit_merge = edit_merge.astype(dtype= {"pageid":"int64",
        "eid":"int64","uid":"int64","timestamp":"object",
        "flags":"bool","sizechange":"int64","comment":"object",
        "diff":"object","inline":"object"})
```

```python
edit_merge.to_csv("E:/study/aml/appaisal project/edit_merge.csv",index=False, header=True)
```

```python
print("The number of rows:",edit_merge.shape[0])
print("The number of rows with minor tags:",edit_merge[edit_merge['flags']==True].shape[0])
```

```python
## partition for minor
edit_minor = edit_merge[edit_merge['flags']==True]
edit_minor.to_csv("E:/study/aml/appaisal project/edit_minor.csv",index=False, header=True)
```
