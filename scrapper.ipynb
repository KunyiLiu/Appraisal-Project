```python
import re

def GetRevisions(pageid):
    url = name + "/w/api.php?action=query&format=xml&prop=revisions&rvlimit=max&pageids=" + quote(pageid) +"&rvprop=ids|flags|timestamp|userid|size|comment"    
    revisions = []                                        #list of all accumulated revisions
    next = ''                                             #information for the next request
    pattern = "<rev revid=\"(.*?)\" parentid=\"(.*?)\" (.*?)userid=\"(.*?)\" timestamp=\"(.*?)\" size=\"(.*?)\" comment=\"(.*?)(?=\")"
    while True:
        response = urlopen(url + next).read()     #web request
        response = response.decode('utf-8')#TypeError: cannot use a string pattern on a bytes-like object print(len(revisions))
        revisions += re.findall(pattern, response)  #adds all revisions from the current request to the list

        cont = re.search('<continue rvcontinue="([^"]+)"', response)
        if not cont:                                      #break the loop if 'continue' element missing
            break

        next = "&rvcontinue=" + cont.group(1)             #gets the revision Id from which to start the next request
    
    ## modify the original extracted info
    re_id, re_pid,re_m,re_user,re_time,re_size,re_comment = zip(*revisions)
    re_m = [ True if x!='' else False for x in re_m] # if it is minor 
    re_sizechange = [int(i)-int(j) for i,j in zip(re_size[:-1],re_size[1:])]
    re_sizechange = list(map(int, re_sizechange))
    re_sizechange.append(int(re_size[-1]))
     
    revisions = list(zip(re_id,re_pid,re_m,re_user,re_time,re_sizechange,re_comment))
    
    return revisions
```

```python
def GetContent(title,re_id, re_pid):
    noDupes = []
    inline = []
    diff = []
    if int(re_pid) == 0:
        ## major points / Criticism
        section = ["Major Points","Criticisms"]
        for i in section:
            url = name + "/w/api.php?action=query&format=xml&prop=revisions&revids=" + quote(re_id) + "&rvprop=content"
            print(url)
            response = urlopen(str(url)).read()
            response = response.decode('utf-8')
            pattern = "==%s== \n*(.+?)\n\n=="%i
            part = re.findall(pattern,response,re.S|re.M)
            if part:
                diff.append('+')
                diff.append("==%s=="%i)
                diff.append(part[0])
        ## concantenate ?
        diff = '\n'.join(diff)
    else:
        request = name + "/w/index.php?title=" + quote(title) + "&diff=" +  quote(re_id)+"&oldid="+quote(re_pid)
        print(request)
        response = urlopen(request)
        wikiHtml = response.read().decode('utf-8')
        html = BeautifulSoup(str(wikiHtml),"lxml")
        ## use beautifulsoup4 
        tag_start = html.find('td',class_= 'diff-lineno')
        if tag_start:
            diff = tag_start.find_all_next('td',class_ =re.compile("diff-lineno|diff-context|diff-addedline"))
            for i in range(len(diff)):
                if diff[i].get('class') == ['diff-context']:
                    if diff[i-1].string != diff[i].string and diff[i].string != None:
                        noDupes.append(diff[i].string)
                elif diff[i].get('class') == ['diff-addedline']:
                    noDupes.append('+')
                    noDupes.append('  **  '.join([x for x in diff[i].stripped_strings]))
                    prep = diff[i].find_all("ins", class_ = "diffchange diffchange-inline")
                    inline+= ([x.string for x in prep])
                    #inline.append(tuple([x.string for x in diff[i].find_all("ins", class_ = "diffchange diffchange-inline")]))
                else:
                    if diff[i-1].get('class') == ['diff-lineno']:
                        noDupes.append(diff[i].string)
        ## concantenate ?
        diff = '\n'.join(noDupes)
    
    return diff,inline
    
```
